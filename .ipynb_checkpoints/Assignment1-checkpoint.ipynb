```python
import numpy as np
from operator import itemgetter
from urllib import request
from nltk.tokenize import word_tokenize
import re
import matplotlib.pyplot as plt
import math
#get corpus of words:
url="http://www.gutenberg.org/files/74/74-0.txt"
reponse=request.urlopen(url)
raw=reponse.read().decode('utf8')
#tokenize words from the corpus:
tokens=word_tokenize(raw)
newTokens=[]
for token in tokens:
    str=''.join(token)
    temp=re.match(r'[A-Za-z]*', str)
    if(temp.group(0)):
        newTokens.append(str)
numtokens=len(newTokens)
print("Number of tokens:",numtokens)
#type from the tokens(unique word list):
types = sorted(set(newTokens))
numtypes = len(types)
print("Number of unique words/types:",numtypes)
#calculating type token ratio:
ttr=float(numtypes/numtokens)
print("Type token ratio:",ttr)
```

```python
#calculating frequency of tokens and creating dictionary:
frequency={}
for word in newTokens:
    if word not in frequency:
        frequency[word] = 1
    else:
        frequency[word] += 1
a={}
rank=1
xaxis=[]
yaxis=[]
zaxis=[]
for key, value in reversed(sorted(list(frequency.items()), key = itemgetter(1)) ):
    if key not in a:
              rank=rank+1
    if(value not in yaxis):
        xaxis.append(rank-1)
        yaxis.append(value)
        zaxis.append(len(key))
#Zipf's Law
plt.scatter(yaxis,zaxis,linewidth=2,color='b')
plt.xlabel("Freqency of a words")
plt.ylabel("Length of words")
plt.title("Zipf's Law showing relationship between the frequency of a word and its corresponding length")
plt.show()
plt.plot(xaxis, yaxis, linewidth=2, color='r')
plt.xlabel("Freqency of a words")
plt.ylabel("Rank of words")
plt.title("Zipf's Law showing relationship between the frequency of a word and its position in the list i.e. its rank")
plt.show()
plt.plot(np.log(xaxis),np.log(yaxis), linewidth=2, color='g')
plt.xlabel("Log of Freqency of a words")
plt.ylabel("Log of Rank of words")
plt.title("Zipf's Law showing relationship between log of frequency of a word and log of its position in the list i.e. log(rank)")          
plt.show()
l1=[]
l1.append(frequency.get('service'))
l1.append(frequency.get('jam'))
l1.append(frequency.get('right'))
l1.append(frequency.get('chapter'))
l1.append(frequency.get('house'))
l2=[7,9,10,7,12]
plt.plot(l1, l2, linewidth=2, color='b')
plt.xlabel("Frequency of words")
plt.ylabel("No. of meanings")
plt.title("Zipf's Law showing relationship between the frequency of a word (f) and its different meanings")
plt.show()
```

```python
def plotting(t=[], length=[]):
    plt.plot(t, length, linewidth=2, color='r')

t = 1
x=[]
y=[]
words = set()
for word in newTokens:
    words.add(word)#unique words i.e. types
    t=t+1
    x.append(t)
    y.append(len(words))
plotting(x,y)
plt.xlabel("No.of tokens")
plt.ylabel("No. of types")
plt.title("Heap's Law showing how the size of the overall vocabulary(number of unique words) grow with the size of the corpus")
plt.show()
xl=[]
yl=[]
for i in range(len(x)):
    if(x[i] is not 0):
        if(y[i] is not 0):
            xl.append(math.log(x[i]))
            yl.append(math.log(y[i]))
slope, intercept = np.polyfit(xl, yl, 1)
print("The values of beta and k according to |V| = KN**β are:")
print(slope," ", math.exp(intercept))
```
